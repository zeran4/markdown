#### 방법 #1 (거듭제곱의 지수가 너무 큰 지 math.e 값과 차이가 큰 듯 ㅠㅠ)

```python
h = 1.0
for n in range(0, 20): #(0.1, 0.01, 0.001, 0.0001, 0.00001, 0.000001, 0.0000001, 0.00000001, 0.000000001):
    h = h * 0.1
    h2 = 1.0 / h
    print('h값 : ', h, "  e값 : ", (1.0 + h)**h2)
    print('1/h값 : ', h2)
```

##### 방법 #1-2(np.float64로 시도 : 다른 방법을 찾아야 할 듯)

```python
h = np.float64(1.0)
for n in range(0, 20): #(0.1, 0.01, 0.001, 0.0001, 0.00001, 0.000001, 0.0000001, 0.00000001, 0.000000001):
    h = h / np.float64(10.)
    print('h값 : ', h, "  e값 : ", (np.float64(1.0) + h)**(np.float64(1.0)/h))
```

#### 방법 2

```python
sum_v = 0.
fact_k = 1. # 기존 factorial 값을 재사용
for k in range(0, 51):
    if k != 0:
        fact_k = fact_k * k
    sum_v = sum_v + (1.0 / fact_k)
    if k % 5 == 0:
        print("k값 : ", k, "  e값 : ", sum_v)
```
